**Part 1:**  <span style="color: rgb(45, 59, 69); font-family: LatoWeb, &quot;Lato Extended&quot;, Lato, &quot;Helvetica Neue&quot;, Helvetica, Arial, sans-serif; font-size: 16px; background-color: rgb(255, 255, 255);">&nbsp;Write a query that finds the total number of sales for the year 2015. Your query cannot use any aggregates. All calculations should be done using the data/time functions in this week's learning. Your output should be in two columns-- year_sales and num_of_sales.</span>

```sql
SELECT 
    EXTRACT(YEAR FROM ord_date) AS year_sales, 
    COUNT(ord_id) AS num_of_sales
FROM 
    dsci_504.Orders
WHERE 
    EXTRACT(YEAR FROM ord_date) = 2015
GROUP BY 
    year_sales;
```

<span style="color: rgb(45, 59, 69); font-family: LatoWeb, &quot;Lato Extended&quot;, Lato, &quot;Helvetica Neue&quot;, Helvetica, Arial, sans-serif; font-size: 16px; background-color: rgb(255, 255, 255);">Find the monthly sales for the same year. use the same output as above</span>

```sql
SELECT 
    TO_CHAR(ord_date, 'YYYY-MM') AS year_month,
    COUNT(*) AS num_of_sales
FROM 
    dsci_504.Orders
WHERE 
    EXTRACT(YEAR FROM ord_date) = 2015
GROUP BY
    TO_CHAR(ord_date, 'YYYY-MM')
ORDER BY 
    year_month;
```

**Part 5:** Write a query that will aggregate the total of all bikes sold. Group the output by the bike manufacturer. Your output should contain the following elements:

- Bike Manufacturer
- Bike Name
- Bike Description
- Bike Country of Origin
- Total of all sales

```sql
SELECT 
    p.prod_manufacturer AS "Bike Manufacturer", 
    p.prod_name AS "Bike Name", 
    p.prod_description AS "Bike Description", 
    p.country_origin AS "Bike Country of Origin", 
    SUM(o.order_tot) AS "Total of all sales"
FROM 
     dsci_504. Products p 
	INNER JOIN   dsci_504.Orders o ON p.prod_id = o.prod_id
GROUP BY 
    p.prod_manufacturer, p.prod_name, p.prod_description, p.country_origin;
```

<span style="color: rgb(45, 59, 69); font-family: LatoWeb, &quot;Lato Extended&quot;, Lato, &quot;Helvetica Neue&quot;, Helvetica, Arial, sans-serif; font-size: 16px; background-color: rgb(255, 255, 255);">Sort your output by the sum of the total order, manufacturer, and the bike name</span>

```sql
SELECT 
    Products.prod_manufacturer AS "Bike Manufacturer", 
    Products.prod_name AS "Bike Name",
    Products.prod_description AS "Bike Description",
    Products.country_origin AS "Bike Country of Origin",
    SUM(Orders.order_tot) AS "Total Sales"
FROM 
     dsci_504.Products
INNER JOIN 
     dsci_504.Orders ON Products.prod_id = Orders.prod_id
GROUP BY 
    Products.prod_manufacturer,
    Products.prod_name,
    Products.prod_description,
    Products.country_origin
ORDER BY 
    "Total Sales" DESC, 
    "Bike Manufacturer" ASC, 
    "Bike Name" ASC;
```

You can include any other columns you feel necessary</span>

```sql
SELECT
  AVG(P.prod_price) AS Average_Bike_Price,
  SUM(O.order_tot) AS Total_Sales,
  (P.prod_price - O.order_tot) AS Price_Difference,
  C.comp_name AS Component_Name
FROM 
  dsci_504.Products P 
  JOIN dsci_504.Orders O ON P.prod_id = O.prod_id
  JOIN dsci_504.Components C ON C.comp_id = P.prod_id
GROUP BY 
  C.comp_name,p.prod_price, o.order_tot;
```
